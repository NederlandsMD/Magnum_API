```python
# Store latitude and longitude in locations
locations = number_of_hospitals_df[["Center Point Lat", "Center Point Lng"]]

# Fill NaN values and convert to float
hospital_density = number_of_hospitals_df["ID"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=hospital_density, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
