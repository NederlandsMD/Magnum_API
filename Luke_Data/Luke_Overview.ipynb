```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Population_Drugs ="Drug_Data.csv"
Drugs_pd = pd.read_csv(Population_Drugs)
Drugs_pd
Final = Drugs_pd[["State or Juristiction ", "Percent distribution of facilities ", "Percent distribution of clients", "Percent of Pop in Rehab"]]
Final_renamed=Final.rename(columns={"State or Juristiction ": "State",
                                    "Percent distribution of facilities ": "Percentage of Drug Rehabilitation Facilities",
                                                "Percent distribution of clients": "Percentage of Patients in Drug Rehabilitation",
                                   "Percent of Pop in Rehab": "Percentage of Population in Rehabilitation"})
Final_renamed.head()
```

```python
Overdose="Overdose.csv"
Overdose_pd = pd.read_csv(Overdose)
Overdose_pd.head()
OverdoseFind=Overdose_pd.rename(columns={"Location": "State", "Opioid Overdose Death Rate (Age-Adjusted)":
                                         "Opioid Overdose Death Rate (Age-Adjusted)", 
                                          "All Drug Overdose Death Rate (Age-Adjusted)":
                                         "All Drug Overdose Death Rate (Age-Adjusted)",
                                         "Percent Change in Opioid Overdose Death Rate from Prior Year":
                                         "Percentage Change in Opioid Overdose Death Rate from Prior Year",
                                         "Percent Change in All Drug Overdose Death Rate from Prior Year":
                                         "Percentage Change in All Drug Overdose Death Rate from Prior Year"})
OverdoseFinal=OverdoseFind[["State", "Percentage Change in Opioid Overdose Death Rate from Prior Year", 
                            "Percentage Change in All Drug Overdose Death Rate from Prior Year"]]
OverdoseFinal.head()
```

```python
Census="nst-est2017-alldata.csv"
Census_df = pd.read_csv(Census)
Data = Census_df[["NAME", "Percent"]]
Data_renamed=Data.rename(columns={"NAME": "State",
                                                "Percent": "Percentage of Total Deaths in the US"
                                                })
Data_renamed.head()
```

```python
Money_pd = pd.read_csv("Population.csv")
Money_pd
State = Money_pd[["State", "Percent of US Population", "Medical Marijuana"]]
State_renamed=State.rename(columns={"State": "State", "Percent of US Population":
                                    "Percentage of US Population"})
State_renamed
```

```python
merge_table1 = pd.merge(Final_renamed, OverdoseFinal, on="State", how="left")
merge_table2 = pd.merge(merge_table1, State_renamed, on="State", how="left")
Final_merge = pd.merge(merge_table2, Data_renamed, on="State", how="left")
Final_merge.head()
```

```python
Data_for_graph = Final_merge[["State", 
                              "Percentage of Drug Rehabilitation Facilities",
                              "Percentage of Patients in Drug Rehabilitation",
                              "Percentage of Population in Rehabilitation",
                              "Percentage Change in Opioid Overdose Death Rate from Prior Year",
                              "Percentage Change in All Drug Overdose Death Rate from Prior Year",
                              "Percentage of US Population",
                              "Percentage of Total Deaths in the US"]]
Data_for_graph.head()
```

```python
Drugs="drugtreatmentcosts.csv"
Drugs = pd.read_csv(Drugs)
Drugs["% of Populaton in Treatment Facilties"] = Data_for_graph["Percentage of Population in Rehabilitation"]
Drugs.head()
```

```python
Data1=Data_for_graph[[ "State","Percentage Change in Opioid Overdose Death Rate from Prior Year",
                              "Percentage Change in All Drug Overdose Death Rate from Prior Year",
                              ]]
Data1.set_index('State', inplace=True)
data_chart = Data1.plot(kind='bar',figsize=(20, 10))
plt.title("Opioid and Drug Overdose Rates in Comparison to the Previous Year")
data_chart.set_xlabel("State")
data_chart.set_ylabel("Percentage Chance Since Previous Year")
data_chart.legend(loc="upper right")
plt.savefig("Opioids and Drugs")
plt.show()
```

```python
Data2=Data_for_graph[[ "State","Percentage of US Population",
                              "Percentage of Total Deaths in the US",
                              ]]
Data2.set_index('State', inplace=True)
data_chart = Data2.plot(kind='bar',figsize=(20, 10))
data_chart.set_xlabel("State")
data_chart.set_ylabel("Data")
data_chart.legend(loc="upper right")
plt.savefig("Death")
plt.show()
```

```python
Data3=Data_for_graph[["State", 
                              "Percentage of Population in Rehabilitation"]]
Data3.set_index('State', inplace=True)
data_chart = Data3.plot(kind='bar',figsize=(20, 10))
plt.title("Population of State in Rehabilitation Programs")
data_chart.set_xlabel("State")
data_chart.set_ylabel("Percentage of US Population")
data_chart.legend(loc="upper right")
plt.savefig("Percent_of_Pop_in_Rehab.png")
plt.show()
```

```python
Data4=Data_for_graph[["State", "Percent of Pop in Rehab", "Percentage of US Population"]]
Data4.set_index('State', inplace=True)
data_chart = Data4.plot(kind='bar',figsize=(20, 10))
data_chart.set_xlabel("State")
data_chart.set_ylabel("Data")
data_chart.legend(loc="upper right")
plt.show()
```
